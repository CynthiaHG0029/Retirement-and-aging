```python
d = pd.DataFrame()

for col in enif_rango.columns[:6]:
  temp = enif_rango.groupby(col).sum().reset_index()
  temp.insert(0, 'id_pregunta',col)
  temp.rename(columns={col:'r_pregunta'}, inplace=True)
  temp = temp[['id_pregunta', 'r_pregunta', 'FAC_ELE']]
  d = pd.concat([d, temp])

d
```

```python
d
```

```python
alt.Chart(d).mark_bar().encode(
  x='id_pregunta',
  y='FAC_ELE',
  color='r_pregunta:N',
  tooltip='FAC_ELE'
)
```
